### Converting a linear regression formula into logistic regression

Let's imagine we already have the needed linear regression formula, and we try to convert this into the logistic regression -format.

Note: this is an over-simplification of how logistic regression works, but the basic principle is there. 

```python
import math
```

```python
# determine the x-variable
x = -4.96

# fit in the linear regression formula 13.56 + 2.73 * x
# into the sigmoid function
linear_regression = 13.56 + 2.73 * x

# put the equation through the sigmoid-function
result = 1 / (1 + math.e ** ((-1) * linear_regression))

# raw version
print(result)

# round to decision 0/1
result = int(round(result, 0))

# inform the user of the actual output
print(f"Final result rounded (0 = no, 1 = yes): {result}")
```
